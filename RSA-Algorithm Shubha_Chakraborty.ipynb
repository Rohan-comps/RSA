```python
import math
import random
 
print("RSA Encryption and Decryption Model")
print("-----------------------------------------------------")

print("Enter the two prime numbers below:")
p = int(input("Enter the first prime number:"))
q = int(input("Enter the second prime number:"))
print("-----------------------------------------------------")
 
    
die = random.SystemRandom()

def single_test(n, a):
    exp = n - 1
    while not exp & 1:
        exp >>= 1
        
    if pow(a, exp, n) == 1:
        return True
        
    while exp < n - 1:
        if pow(a, exp, n) == n - 1:
            return True
            
        exp <<= 1
        
    return False
    
def miller_rabin(n, k=40):
    for i in range(k):
        a = die.randrange(1, n - 1)
        if not single_test(n, a):
            return False
            
    return True
 
check_p = miller_rabin(p)
check_q = miller_rabin(q)
while(((check_p==False)or(check_q==False))):
    p = int(input("Enter the first prime number:"))
    q = int(input("Enter the second prime number:"))
    check_p = miller_rabin(p)
    check_q = miller_rabin(q)

n = p * q
print("RSA Modulus(n) is:",n)
 
r= (p-1)*(q-1)
print("Eulers Toitent(r) is:",r)
print("-----------------------------------------------------")
 
def egcd(e,r):
    while(r!=0):
        e,r=r,e%r
    return e
 
#Euclid's Algorithm
def eugcd(e,r):
    for i in range(1,r):
        while(e!=0):
            a,b=r//e,r%e
            r=e
            e=b
 
#Extended Euclidean Algorithm
def eea(a,b):
    if(a%b==0):
        return(b,0,1)
    else:
        gcd,s,t = eea(b,a%b)
        s = s-((a//b) * t)
        return(gcd,t,s)
 
#Multiplicative Inverse
def mult_inv(e,r):
    gcd,s,_=eea(e,r)
    if(gcd!=1):
        return None
    else:
        return s%r
    
e = int(input("Enter a prime number for e: "))
if(egcd(e,r)==1) & e<=r-1 & e>=2:
    print("The value of e is:",e)
print("-----------------------------------------------------")
eugcd(e,r)
d = mult_inv(e,r)
print("The value of d is:",d)
print("-----------------------------------------------------")
public = (e,n)
private = (d,n)
print("Private Key is:",private)
print("Public Key is:",public)
print("-----------------------------------------------------")

def encrypt(pub_key,m):
    e,n=pub_key
    c = (m**e)%n
    return c
     
def decrypt(priv_key,c):
    d,n=priv_key
    m = (c**d)%n
    return m
 
choose = input("Press '1' for encryption and '2' for decryption :")
if(choose=='1'):
    message = int(input("Enter the message:"))
    enc_msg=encrypt(public,message)
    print("Your encrypted message is:",enc_msg)
elif(choose=='2'):
    cipher = int(input("Enter the cipher test:"))
    print("Your decrypted message is:",decrypt(private,cipher))
else:
    print("You entered the wrong option.")
```
